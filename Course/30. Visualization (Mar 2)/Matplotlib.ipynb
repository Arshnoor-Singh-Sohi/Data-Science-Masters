```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10, 200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xlabel("This is my X data")
plt.ylabel("This is my Y data")
plt.title("This is my title")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
colours = np.random.rand(50)
sizes = 500 * np.random.rand(50)
plt.scatter(x,y, c = colours, s = sizes, alpha = .8)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Random Data")
```

```python
x = ['a', 'b', 'c', 'd', 'e']
```

```python
y = np.random.rand(5)
plt.bar(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Bar Plot")
y
```

```python
plt.barh(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Bar Plot")
```

```python
plt.figure(figsize=(8,4))
plt.plot(x, y, '--r')
plt.xlabel("This is my X data")
plt.ylabel("This is my Y data")
plt.title("This is my title")
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colours = np.random.rand(50)
sizes = 500 * np.random.rand(50)
plt.scatter(x,y, c = 'b', s = sizes, alpha = .8, marker = 'v')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Random Data")
```
