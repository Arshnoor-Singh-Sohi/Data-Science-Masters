## Covariance and Correlation

```python
import seaborn as sns
```

```python
df = sns.load_dataset('healthexp')
```

```python
df.head()
```

```python
## Covariance
df.cov()
```

```python
## Pearson Correlation Coefficient
df.corr(method = 'pearson')
```

```python
## Spearman Rank Correlation
df.corr(method = 'spearman')
```
